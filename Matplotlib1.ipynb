# Line Plot

A line chart or line plot is a graphical representation used to show the relationship between two continuous variables by connecting data points with a straight line. It is commonly used to visualize trends, patterns or changes over time.

In Matplotlib line charts are created using the pyplot sublibrary which provides simple and flexible functions for plotting data. In a line chart, the x-axis typically represents the independent variable while the y-axis represents the dependent variable.



**Syntax: plt.plot(x, y, color, linestyle, linewidth, marker)**

### Adding Labels and Title to a Line Plot

```python

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temperature = [22, 24, 23, 26, 25]

plt.plot(days,temperature,marker = 'o')
plt.xlabel('Days')
plt.ylabel('Temperature(C)')
plt.title('Weekly temperature')
plt.show()
```

### Adding Grid to a Line Chart

Grids make it easier to read values and follow trends across the chart. You can enable grids using the grid() function

| Parameter   | Description     |
| ----------- | --------------- |
| `True`      | Show grid       |
| `False`     | Hide grid       |
| `color`     | Grid color      |
| `linestyle` | Grid line style |
| `linewidth` | Thickness       |


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.plot(x,y,marker='o')

plt.grid(
    True,
    color='gray',
    linestyle='--',
    linewidth=1
)

plt.show()
```

### Annotations

For adding annotations to a line chart you can use the annotate() function. This function allows you to display additional information such as the exact x and y values directly on the data points, improving clarity and data interpretation.

| Parameter    | Meaning             |
| ------------ | ------------------- |
| `xy`         | Actual data point   |
| `xytext`     | Text position       |
| `arrowprops` | Arrow customization |

for xi, yi in zip(x, y):

    )

```python
from matplotlib.typing import ColourType
years = [2018,2019,2020,2021,2022,2023,2024,2025]
scholarships = [120,150,180,250,320,400,480,360]
plt.figure(figsize=(12,10))

plt.plot(years,scholarships,marker='o',label='Scholarship')
plt.title('Schoolarship record across country')
plt.xlabel('Years')
plt.ylabel('Schoolarship')

for xi,yi in zip (years,scholarships):
  plt.annotate(f'{xi},{yi}', (xi,yi),
               textcoords='offset points',
               xytext=(0,10),
               ha='center')

plt.grid(True)

plt.show()

```

### Multiple Line Charts Using Matplotlib

To create multiple line charts in separate containers you can use the figure() function. Each call to figure() generates a new plotting area, making it easier to visualize and compare different datasets independently.

```python
months = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]
service_revenue = [
    1200, 1400, 1600, 1800, 2100, 2400,
    2600, 2800, 3000, 3300, 3500, 3800
]
sales_revenue = [
    2000, 2300, 2500, 2700, 3000, 3400,
    3600, 3900, 4200, 4500, 4700, 5000
]

plt.plot(months,service_revenue,marker='o',linewidth=3,color='green',label='service_revenue')
plt.plot(months,sales_revenue,marker='o',linewidth=3,color='blue',label='sales_revenue')

plt.grid(True)
plt.title('Revenue')
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

### Fill the Area Between Two Lines

To fill the region between two line plots, you can use the fill_between() function. This function shades the area between two curves, helping visualize the difference or range between datasets.

legend() in Matplotlib:

***legend() shows labels for plotted lines***

```python
months = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]
service_revenue = [
    1200, 1400, 1600, 1800, 2100, 2400,
    2600, 2800, 3000, 3300, 3500, 3800
]
sales_revenue = [
    2000, 2300, 2500, 2700, 3000, 3400,
    3600, 3900, 4200, 4500, 4700, 5000
]

plt.plot(months,service_revenue,marker='o',linewidth=3,color='green',label='service_revenue')
plt.plot(months,sales_revenue,marker='o',linewidth=3,color='blue',label='sales_revenue')

plt.grid(True)
plt.fill_between(months,service_revenue,sales_revenue,color='orange',alpha=0.4)
plt.title('Revenue')
plt.xlabel('Months')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

### Plotting Trigonometric Functions

Trigonometric functions such as sine, cosine and tangent can be visualized using plt.plot() by generating input values over a continuous range. The numpy.linspace() function creates evenly spaced values, allowing smooth and accurate curves.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")
plt.xlabel("X values")
plt.ylabel("Function value")
plt.title("Trigonometric Functions")
plt.legend()
plt.show()
```

# Bar Plot

A bar plot uses rectangular bars to represent data categories, with bar length or height proportional to their values. It compares discrete categories, with one axis for categories and the other for values.

**Why Use Bar Plots?**

Bar plots are significant because they provide a clear and intuitive way to visualize categorical data. They allow viewers to quickly grasp differences in size or quantity among categories, making them ideal for presenting survey results, sales data, or any discrete variable comparisons.

```python

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

plt.bar(fruits, sales)
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

## Multiple Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

coffee_sales = [1200, 1500, 1700, 2000, 2200, 2500]

tea_sales = [1000, 1300, 1500, 1700, 1900, 2100]

honey_sales = [800, 950, 1100, 1300, 1500, 1800]

x = np.arange(len(months))

width = 0.25

plt.figure(figsize=(12,8))

# Coffee Bars
plt.bar(
    x - width, #left side shifting
    coffee_sales,
    width,
    label='Coffee'
)

# Tea Bars
plt.bar(
    x, # middle side shifting
    tea_sales,
    width,
    label='Tea'
)

# Honey Bars
plt.bar(
    x + width,#right side shifting
    honey_sales,
    width,
    label='Honey'
)

plt.xticks(x, months)

plt.xlabel('Months')

plt.ylabel('Sales Amount')

plt.title('Coffee vs Tea vs Honey Sales Report')

plt.grid(True)

plt.legend()

plt.show()
```

A stacked bar plot places bars ON TOP of each other.

**Used for:**

showing total + category

contribution

comparing composition

sales breakdown

revenue analysis

Side by Side

One on top of another

**Basic Syntax**

plt.bar(x, data1)

plt.bar(x, data2, bottom=data1)

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

coffee = [120, 150, 170, 200, 220]

tea = [100, 120, 140, 160, 180]

honey = [80, 90, 100, 120, 140]

plt.figure(figsize=(10,5))

# Coffee
plt.bar(
    months,
    coffee,
    label='Coffee'
)

# Tea stacked on Coffee
plt.bar(
    months,
    tea,
    bottom=coffee,
    label='Tea'
)

# Honey stacked on Coffee + Tea
plt.bar(
    months,
    honey,
    bottom=[
        coffee[i] + tea[i]
        for i in range(len(coffee))
    ],
    label='Honey'
)

plt.title('Coffee vs Tea vs Honey Sales')

plt.xlabel('Months')

plt.ylabel('Sales')

plt.grid(True)

plt.legend()

plt.show()
```
