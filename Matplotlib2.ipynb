# Histogram

**A histogram is used to show the frequency distribution**

 of numerical continuous data.

It groups data into intervals called "bins"

 and shows how many values fall into each interval.

 **Histogram helps understand:**
 - Data distribution
 - Most common value ranges
 - Spread of data
 - Skewness and patterns

IMPORTANT PARAMETERS
# ------------------------------------------

  bins       -> Number of intervals/groups

 color      -> Bar color


 edgecolor  -> Border color of bars

 alpha      -> Transparency level

 **Histogram:**
 - Used for numerical continuous data
 - Shows frequency distribution
 - Bars touch each other

 **Bar Plot:**
 - Used for categorical data
 - Compares categories
 - Bars usually separated

## Basic Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.rand(1000)
plt.hist(data,bins=40,color='skyblue',edgecolor='black')
plt.xlabel('Values')
plt.ylabel('frequenscy')
plt.title('Basic Histogram')

plt.show()
```

Explanation:

Generates 1000 random numbers from a standard normal distribution.

Plots a histogram with 30 bins, sky-blue bars and black edges.

Adds X and Y axis labels and a title.

Displays the histogram plot.

## Customized Histogram with Density Plot

```python
import seaborn as sns
# Generate random data for the histogram
data = np.random.randn(1000)

# Creating a customized histogram with a density plot
sns.histplot(data, bins=30, kde=True, color='lightgreen', edgecolor='red')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Customized Histogram with Density Plot')

# Display the plot
plt.show()
```

Explanation:

Generates 1000 random numbers.


Plots a histogram with 30 bins and a smooth density curve (KDE) using Seaborn.

Colors bars light green with red edges.

Adds axis labels and a title.

Displays the plot showing data distribution and density.

## Multiple histogram

```python
import matplotlib.pyplot as plt

coffee = [1200,1300,1400,1500,1600]

tea = [1000,1100,1200,1300,1400]

honey = [800,900,1000,1100,1200]

plt.hist(
    coffee,
    bins=5,
    alpha=0.5,
    label='Coffee'
)

plt.hist(
    tea,
    bins=5,
    alpha=0.5,
    label='Tea'
)

plt.hist(
    honey,
    bins=5,
    alpha=0.5,
    label='Honey'
)

plt.legend()

plt.show()
```

# Scatter Plot

Scatter plots are one of the most fundamental tools for visualizing relationships between two numerical variables. matplotlib.pyplot.scatter() plots points on a Cartesian plane defined by X and Y coordinates. Each point represents a data observation, allowing us to visually analyze how two variables correlate, cluster or distribute.

image.png

image.png

```python
import matplotlib.pyplot as plt

study_hours = [1,2,3,4,5]

marks = [40,50,60,70,85]

plt.scatter(
    study_hours,
    marks
)

plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
import matplotlib.pyplot as plt

advertisement = [
    1000,1500,2000,2500,3000,
    3500,4000,4500
]

sales = [
    5000,6500,8000,9000,
    11000,12500,14000,15500
]

plt.figure(figsize=(10,5))

plt.scatter(
    advertisement,
    sales,

    color='blue',

    s=90
)

plt.title('Advertisement vs Sales')

plt.xlabel('Advertisement Cost ($)')

plt.ylabel('Sales Revenue ($)')

plt.grid(True)

plt.show()
```

# Pie chart

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.pie(data, labels=cars)

# show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

products = ['Coffee', 'Tea', 'Honey', 'Milk']

sales = [45, 30, 15, 10]

colors = ['brown', 'green', 'gold', 'skyblue']

explode = [0.1, 0, 0, 0]

plt.figure(figsize=(8,8))

plt.pie(
    sales,

    labels=products,

    colors=colors,

    autopct='%1.1f%%',

    explode=explode,

    shadow=True,

    startangle=90
)

plt.title('Product Sales Contribution')

plt.show()
```

# HeatMap

A heatmap is a data visualization technique that uses color-coded intensity to represent magnitudes of data, making complex trends instantly clear

```python
import matplotlib.pyplot as plt

marks = [
    [80,85,90],
    [70,75,80],
    [60,65,70]
]

plt.figure(figsize=(6,5))

plt.imshow(
    marks,
    cmap='hot'
)

plt.colorbar()

plt.title('Student Marks Heatmap')

plt.xlabel('Subjects')

plt.ylabel('Students')

plt.show()
```

# Box Plot

A Box Plot (or Whisker plot) display the summary of a data set, including minimum, first quartile, median, third quartile and maximum. it consists of a box from the first quartile to the third quartile, with a vertical line at the median. the x-axis denotes the data to be plotted while the y-axis shows the frequency distribution. The matplotlib.pyplot module of matplotlib library provides boxplot() function with the help of which we can create box plots.

**Syntax:**

 matplotlib.pyplot.boxplot(data)

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)

d_1 = np.random.normal(100, 10, 200)
d_2 = np.random.normal(90, 20, 200)
d_3 = np.random.normal(80, 30, 200)
d_4 = np.random.normal(70, 40, 200)
d = [d_1, d_2, d_3, d_4]

fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(d)

plt.show()
```
